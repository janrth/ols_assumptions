```python
import olscheck
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# instantiate OLS assumptions check:

ols_checker = olscheck.OlsCheck()
```

```python
# generate data:

data_gen = olscheck.data_generator.LinearRegressionDataGenerator()

df = data_gen.generate_data()
```

```python
# fit OLS:

X = df[['X1','X2']]
y = df.y.values

model = LinearRegression(fit_intercept=True)
model.fit(X, y)

df['pred'] = model.predict(X)
```

```python
# Create plots:

ols_checker.ols_assumptions(df=df, 
                            y_true_col='y',
                            fittedvalues_col='pred',
                            features=['X1','X2'],
                            axis_lim=True,
                            constant=True)
```
